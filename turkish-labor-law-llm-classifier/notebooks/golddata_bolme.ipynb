```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Veriyi yükle
df = pd.read_excel("drive/MyDrive/LLM-PROJE/gold_data.xlsx")

train_df, temp_df = train_test_split(
    df,
    test_size=0.20,
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42
)

# 4. Dosyaları kaydet
train_df.to_csv("drive/MyDrive/LLM-PROJE/train.csv", index=False)
val_df.to_csv("drive/MyDrive/LLM-PROJE/val.csv", index=False)
test_df.to_csv("drive/MyDrive/LLM-PROJE/test.csv", index=False)

print(f"Eğitim Seti: {len(train_df)} | Doğrulama Seti: {len(val_df)} | Test Seti: {len(test_df)}")
```
